```python
spark.sql("CREATE SCHEMA IF NOT EXISTS mini_project_catalog.`01_bronze`")

files = [
    "customer_survey.csv",
    "estimate.csv",
    "invoice.csv",
    "order.csv",
    "store.csv",
    "ns_budget.csv"
]

s3_base_path = "s3://automobile-data-12345/"   # your bucket

for file in files:
    table_name = file.replace(".csv", "")
    
    df = spark.read.csv(
        s3_base_path + file,
        header=True,
        inferSchema=True
    )
    
    df.write.mode("overwrite") \
      .saveAsTable(f"mini_project_catalog.`01_bronze`.{table_name}")

print("Bronze tables created in 01_bronze successfully!")
```
